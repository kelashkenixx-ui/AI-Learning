```python
print("--- Running Task 3: Multi-Tool Orchestrator ---\n")

# 1. Alag-alag custom tools define karein
def check_weather_tool(city):
    return f"⚡ [Tool Output] Weather in {city} is 28°C, Clear Sky."

def check_time_tool(city):
    return f"🕒 [Tool Output] Current Time in {city} is 01:45 PM."

# 2. Orchestrator Logic jo automatically sahi tool select karti hai
def orchestrator_agent(user_prompt):
    print(f"User Request: '{user_prompt}'")

    if "weather" in user_prompt.lower() or "temperature" in user_prompt.lower():
        print("💡 Orchestrator: Routing to Weather Tool...")
        return check_weather_tool("Karachi")
    elif "time" in user_prompt.lower() or "clock" in user_prompt.lower():
        print("💡 Orchestrator: Routing to Time Tool...")
        return check_time_tool("Karachi")
    else:
        return "Standard response from LLM core."

# Run and check output
final_output = orchestrator_agent("Can you check the weather of Karachi right now?")
print(f"\nFinal System Output:\n{final_output}")
```
