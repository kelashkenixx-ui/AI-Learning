```python
import os
from google import genai
from google.genai import types
from google.colab import userdata

# 1. Direct secure retrieval from your secrets panel
try:
    api_key_val = userdata.get('GEMINI_API_KEY')
    client = genai.Client(api_key=api_key_val)
except Exception as e:
    print("Error loading key. Please check your Colab Secrets (🔑) panel!")

# 2. Strict calculation instruction context
system_prompt = """
You are an expert AI Calculator Agent with perfect conversational memory.
Perform the requested math operations. Always remember the previous calculation result.
If the user says an operation like 'multiply by 2' without a starting number, you must execute that operation on the PREVIOUS result.
"""

print("--- Running Task 2: AI Calculator with Stable Memory Context ---\n")

# --- FIRST INTERACTION ---
prompt1 = "Start with 10 and add 20"
print(f"User Request 1: {prompt1}")

# Passing instructions along with the content directly
config = types.GenerateContentConfig(system_instruction=system_prompt, temperature=0.1)
response1 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt1,
    config=config
)
output1 = response1.text
print(f"Agent Output 1: {output1}\n")

print("-" * 50)

# --- SECOND INTERACTION (Forcing context history to check memory state) ---
prompt2 = f"Based on your previous answer which was {output1}, now multiply that by 2"
print(f"User Request 2: Now multiply that by 2")

response2 = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt2,
    config=config
)
print(f"Agent Output 2 (Memory Verified): {response2.text}\n")
```
