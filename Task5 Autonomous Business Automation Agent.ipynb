```python
from google import genai
from google.colab import userdata

api_key_val = userdata.get('GEMINI_API_KEY')
client = genai.Client(api_key=api_key_val)

print("--- Running Task 5: Autonomous Business Automation Agent ---\n")

incoming_emails = [
    "URGENT: Database cluster is down! Production API is failing.",
    "Hey, do you want to win a free $1000 Amazon gift card? Click this link now!",
    "Hi Kelash, just checking if we have our project status meeting today at 4 PM?"
]

for index, email in enumerate(incoming_emails, 1):
    automation_prompt = f"Classify this email text into exactly one label from [URGENT, SPAM, PERSONAL]. Output only the label in uppercase:\n'{email}'"

    response = client.models.generate_content(
        model='gemini-2.5-flash',
        contents=automation_prompt,
    )
    print(f"📩 Processing Email #{index}: '{email}'")
    print(f"⚙️ Automation Action Taken: [{response.text.strip()}]\n")
```
